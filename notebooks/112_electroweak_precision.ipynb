```python
# ── S1: Two Definitions of sin²θ_W and the Scheme Gap ──
#
# The Standard Model has two common definitions of sin²θ_W:
#   ON-SHELL: sin²θ_W = 1 - M_W²/M_Z²  (physical masses)
#   MS-BAR:   running coupling ratio at scale μ = M_Z
#
# They differ by ~3.6% due to radiative corrections.
# The solenoid gives BOTH:
#   NB30/NB32: 8/35 ≈ 0.22857 (closer to on-shell)
#   NB111:     0.23129          (matches MS-bar)

print("THE SCHEME GAP: On-Shell vs MS-bar sin²θ_W")
print("=" * 65)

# PDG values in both schemes
print(f"\n  PDG on-shell:  sin²θ_W = 1 - M_W²/M_Z² = {sw2_onshell:.5f}")
print(f"  PDG MS-bar:    sin²θ_W(M_Z)             = {sw2_MSbar:.5f}")
print(f"  Scheme gap:    Δ = {sw2_MSbar - sw2_onshell:.5f}")
print(f"  Relative gap:  {(sw2_MSbar - sw2_onshell)/sw2_onshell*100:.2f}%")

# Solenoid predictions
print(f"\n  Solenoid:")
print(f"    NB30 tree:   sin²θ_W = 8/35 = {float(sw2_tree):.5f}")
print(f"    NB111 corr:  sin²θ_W_MS     = {sw2_NB111:.5f}")
print(f"    Solenoid gap: Δ = {sw2_NB111 - float(sw2_tree):.5f}")
print(f"    Relative gap:  {(sw2_NB111 - float(sw2_tree))/float(sw2_tree)*100:.2f}%")

# How close is each to its target scheme?
dev_tree_OS = abs(float(sw2_tree) - sw2_onshell) / sw2_onshell * 100
dev_NB111_MS = abs(sw2_NB111 - sw2_MSbar) / sw2_MSbar_unc  # in sigma
print(f"\n  Accuracy:")
print(f"    |8/35 - on-shell|   = {abs(float(sw2_tree) - sw2_onshell):.5f}  ({dev_tree_OS:.2f}%)")
print(f"    |NB111 - MS-bar|    = {abs(sw2_NB111 - sw2_MSbar):.5f}  ({dev_NB111_MS:.1f}σ)")

# Key question: can we improve M_W?
# M_W from on-shell: M_W = M_Z √(1 - sin²θ_W_OS)
# NB32's prediction uses sin²θ_W = 8/35 → on-shell-like:
MW_from_tree = M_Z * np.sqrt(1 - float(sw2_tree))
# NB111 MS-bar naively applied to on-shell formula:
MW_from_NB111_naive = M_Z * np.sqrt(1 - sw2_NB111)

print(f"\n  M_W predictions:")
print(f"    NB32 (8/35):          M_W = {MW_from_tree:.4f} GeV  ({abs(MW_from_tree-M_W)/M_W_unc:.1f}σ)")
print(f"    NB111 naive (0.23129): M_W = {MW_from_NB111_naive:.4f} GeV  ({abs(MW_from_NB111_naive-M_W)/M_W_unc:.1f}σ)")
print(f"    PDG:                   M_W = {M_W:.4f} GeV")

print(f"\n  NB32's tree-level 8/35 gives BETTER M_W than NB111's MS-bar value!")
print(f"  This is correct: 8/35 is the on-shell prediction, 0.23129 is MS-bar.")
print(f"  The solenoid naturally produces BOTH scheme values.")

# The scheme gap structure
delta_sw2_measured = sw2_MSbar - sw2_onshell
delta_sw2_solenoid = sw2_NB111 - float(sw2_tree)
print(f"\n  SCHEME GAP COMPARISON:")
print(f"    Measured gap:  Δsin²θ_W = {delta_sw2_measured:.5f}")
print(f"    Solenoid gap:  Δsin²θ_W = {delta_sw2_solenoid:.5f}")
print(f"    Ratio: solenoid/measured = {delta_sw2_solenoid/delta_sw2_measured:.3f}")
gap_dev = abs(delta_sw2_solenoid - delta_sw2_measured) / delta_sw2_measured * 100
print(f"    Deviation: {gap_dev:.1f}%")
```

```python
# ── S2: M_W from Different Solenoid Routes ──
#
# Route A: M_W = M_Z cos θ_W (tree-level, NB32)
# Route B: M_W = ½ g₂ v_measured (NB111 corrected g₂ + measured G_F)
# Route C: Full solenoid chain M_Z → M_W → v → G_F

print("M_W FROM DIFFERENT SOLENOID ROUTES")
print("=" * 65)

# Route A: NB32 tree-level
MW_A = M_Z * np.sqrt(float(cw2_tree))
dev_A = (MW_A - M_W) / M_W * 100
sig_A = abs(MW_A - M_W) / M_W_unc

print(f"\n  Route A: M_W = M_Z √(27/35)  [NB32 tree, on-shell sin²θ_W = 8/35]")
print(f"    M_W = {MW_A:.4f} GeV  (dev = {dev_A:+.3f}%, {sig_A:.1f}σ)")

# Route B: NB111 corrected g₂ + measured v
g2_corr = np.sqrt(4 * np.pi / inv_a2_corr)
MW_B = 0.5 * g2_corr * v_EW
dev_B = (MW_B - M_W) / M_W * 100
sig_B = abs(MW_B - M_W) / M_W_unc

print(f"\n  Route B: M_W = ½ g₂(corr) × v(measured)  [NB111 α₂ + PDG G_F]")
print(f"    g₂ = √(4π/29.586) = {g2_corr:.6f}")
print(f"    v  = (√2 G_F)^(-½) = {v_EW:.4f} GeV")
print(f"    M_W = {MW_B:.4f} GeV  (dev = {dev_B:+.3f}%, {sig_B:.1f}σ)")

# Route C: Full solenoid chain — use corrected g₂ but NB32 M_W to get v
MW_C = MW_A   # Tree-level M_W as starting point
v_sol = 2 * MW_C / g2_corr   # Solenoid vev from corrected g₂
GF_sol = 1 / (np.sqrt(2) * v_sol**2)  # Solenoid G_F
dev_v = (v_sol - v_EW) / v_EW * 100
dev_GF = (GF_sol - G_F) / G_F * 100

print(f"\n  Route C: Full solenoid chain [NB32 M_W + NB111 g₂]")
print(f"    M_W = {MW_C:.4f} GeV  [from NB32]")
print(f"    g₂  = {g2_corr:.6f}  [from NB111]")
print(f"    v   = 2M_W/g₂ = {v_sol:.4f} GeV  (dev = {dev_v:+.3f}%)")
print(f"    G_F = 1/(√2v²) = {GF_sol:.7e} GeV⁻²  (dev = {dev_GF:+.3f}%)")

# Route D: Use M_Z and corrected g₂ directly
# M_Z = g_Z v/2 where g_Z² = g₂²/cos²θ_W = g₂² + g'²
# So M_Z² = (g₂² + g'²)v²/4
# And M_W = M_Z × g₂/g_Z = M_Z × g₂/√(g₂²+g'²) = M_Z cos θ_W
# with cos θ_W = g₂/√(g₂²+g'²)
# g' = g₂ tan θ_W
# But this just reduces to Route A again.

# The improvement from A->B and what it means
print(f"\n  IMPROVEMENT from ρ-correction:")
print(f"    Route A (tree):     {dev_A:+.3f}%  ({sig_A:.1f}σ)")
print(f"    Route B (corr+meas): {dev_B:+.3f}%  ({sig_B:.1f}σ)")
print(f"    Improvement: {abs(dev_A)/abs(dev_B):.1f}× better")

# What M_W does Route B predict, and what on-shell sin²θ_W does it imply?
sw2_OS_from_B = 1 - (MW_B/M_Z)**2
print(f"\n  Route B implies sin²θ_W(on-shell) = {sw2_OS_from_B:.5f}")
print(f"  PDG on-shell:                        {sw2_onshell:.5f}")
print(f"  NB32 tree (8/35):                    {float(sw2_tree):.5f}")
```

```python
# ── S3: The Δr Radiative Correction and Its Anatomy ──
#
# The master EW relation: M_W²(1-Δr)sin²θ_W(OS) = πα(0)/(√2 G_F)
# Δr contains the radiative corrections that shift tree → measured.
# Its main components:
#   Δα: running of α from Q=0 to M_Z
#   Δρ: custodial symmetry breaking (mainly from top mass)
#   remainder: Higgs + other loop corrections

print("THE Δr RADIATIVE CORRECTION")
print("=" * 65)

# ── The master relation ──
# πα(0)/(√2 G_F) is a precisely known combination
A_sq = np.pi * alpha_0 / (np.sqrt(2) * G_F)
print(f"\n  Master constant: πα(0)/(√2 G_F) = {A_sq:.4f} GeV²")

# From M_W²(1-Δr)sin²θ_W(OS) = A²:
# Δr = 1 - A²/(M_W² sin²θ_W(OS))
Delta_r_measured = 1 - A_sq / (M_W**2 * sw2_onshell)
print(f"\n  Δr (from PDG values):")
print(f"    Δr = 1 - πα(0)/(√2 G_F M_W² sin²θ_W(OS))")
print(f"    Δr = 1 - {A_sq:.4f}/({M_W**2:.4f} × {sw2_onshell:.5f})")
print(f"    Δr = {Delta_r_measured:.6f}")

# ── Anatomy of Δr ──
# Component 1: Δα — running of α
Delta_alpha = 1 - alpha_0/alpha_MZ
print(f"\n  ANATOMY OF Δr:")
print(f"  ─────────────")
print(f"  Δα = 1 - α(0)/α(M_Z) = 1 - {1/alpha_0:.3f}/{1/alpha_MZ:.3f}")
print(f"     = {Delta_alpha:.6f}")

# Component 2: Δρ — custodial symmetry breaking (top quark)
Delta_rho = 3 * G_F * m_t**2 / (8 * np.pi**2 * np.sqrt(2))
cw2_os = 1 - sw2_onshell
ct_ratio = cw2_os / sw2_onshell
print(f"\n  Δρ = 3G_F m_t²/(8π²√2) = {Delta_rho:.6f}")
print(f"  cos²θ/sin²θ = {ct_ratio:.4f}")
print(f"  −(cos²θ/sin²θ)Δρ = {-ct_ratio * Delta_rho:.6f}")

# Approximate total
Delta_r_approx = Delta_alpha - ct_ratio * Delta_rho
remainder = Delta_r_measured - Delta_r_approx
print(f"\n  Δr ≈ Δα − (c²/s²)Δρ + remainder")
print(f"      ≈ {Delta_alpha:.6f} − {ct_ratio * Delta_rho:.6f} + remainder")
print(f"      ≈ {Delta_r_approx:.6f} + remainder")
print(f"  Remainder: {remainder:.6f}")
print(f"  Total Δr: {Delta_r_measured:.6f}")

# ── What Δr does the solenoid need? ──
# If M_W(sol) = 80.091 (NB32), what Δr would give exact M_W?
# M_W_true² (1-Δr) sin²θ_W = A²
# We want M_W_true = 80.369
# First: what Δr does NB32's prediction imply?
Delta_r_from_NB32 = 1 - A_sq / (MW_A**2 * float(sw2_tree))
print(f"\n  SOLENOID Δr REQUIREMENT:")
print(f"  If sin²θ_W(OS) = 8/35 and M_W = {MW_A:.4f}:")
print(f"    Δr(solenoid tree) = {Delta_r_from_NB32:.6f}")
print(f"    vs Δr(measured)   = {Delta_r_measured:.6f}")
print(f"    Gap: {abs(Delta_r_from_NB32 - Delta_r_measured):.6f}")

# ── Solenoid Δα ──
# In the solenoid, 1/α(M_Z) = 127.919 (NB111, corrected)
# 1/α(0) = 137.036 (PDG, no solenoid prediction yet)
# So: Δα_sol = 1 - 127.919/137.036 = 1 - α_sol(M_Z)/α(0)
# The solenoid would need α(0) to predict Δα
Delta_alpha_sol = 1 - inv_aem_corr / (1/alpha_0)
print(f"\n  Solenoid Δα using NB111 α(M_Z):")
print(f"    Δα = 1 - {inv_aem_corr:.3f}/{1/alpha_0:.3f} = {Delta_alpha_sol:.6f}")
print(f"    PDG Δα = {Delta_alpha:.6f}")
print(f"    Gap: {abs(Delta_alpha_sol - Delta_alpha):.6f}")

# ── Solenoid Δρ — depends on top mass ──
# NB34 tree: y_t = 1/√P₁, so m_t = v/√(2P₁) = v/2
# NB34 m_t was 123.75 GeV (tree) vs 172.69 (measured): 28% off
# The CASCADE mechanism (NB60-108) gives correct mass RATIOS
# but absolute mass would need v correction
m_t_tree = v_EW / (np.sqrt(2*P1) )  # using measured v for fair comparison
Delta_rho_tree = 3 * G_F * m_t_tree**2 / (8 * np.pi**2 * np.sqrt(2))
print(f"\n  Solenoid Δρ with tree top mass:")
print(f"    m_t(tree) = v/√(2P₁) = {m_t_tree:.2f} GeV  (PDG: {m_t} GeV)")
print(f"    Δρ(tree) = {Delta_rho_tree:.6f}  (PDG: {Delta_rho:.6f})")
print(f"    Ratio: tree/measured = {Delta_rho_tree/Delta_rho:.3f}")
print(f"    m_t² ratio: {m_t_tree**2/m_t**2:.3f}  (= 1/(2P₁) × v²/m_t² correction)")

# ── The connection to cascade work ──
print(f"\n  CONNECTION TO LATER WORK:")
print(f"  The cascade mechanism (NB60-108) determines fermion mass")
print(f"  RATIOS from {'{'}2,3,5,7{'}'}. But absolute masses need v.")
print(f"  The top mass is the dominant correction to M_W via Δρ.")
print(f"  Improving m_t(solenoid) → improves Δρ → improves M_W.")
print(f"  This is the bridge between gauge sector and mass sector.")
```
