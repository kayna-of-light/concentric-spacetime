## 6. Spectral Dimension of the Solenoid

The spectral dimension $d_s(t)$ measures the effective dimensionality
of the Cayley graph at diffusion scale $t$:

$$d_s(t) = -\frac{2t\, Z'(t)}{Z(t)}, \quad
Z(t) = \text{Tr}(e^{-tL}) = \sum_k d_k \, e^{-kt}$$

```python
# ── Spectral dimension ──
eigs_flat = []
for k, d in eigen_mult.items():
    eigs_flat.extend([k] * d)
eigs_flat = np.array(eigs_flat, dtype=float)

ts = np.logspace(-2, 2, 500)
ds_vals = np.zeros(len(ts))

for i, t in enumerate(ts):
    w = np.exp(-t * eigs_flat)
    Z  = np.sum(w)
    Zp = -np.sum(eigs_flat * w)
    ds_vals[i] = -2 * t * Zp / Z

peak_idx = np.argmax(ds_vals)
d_peak = ds_vals[peak_idx]
t_peak = ts[peak_idx]

print("SPECTRAL DIMENSION d_s(t)")
print("=" * 70)
print(f"\n  Peak: d_s = {d_peak:.4f} at t = {t_peak:.4f}")

# Report at key scales
key_t = [0.01, 0.1, 0.5, 1.0, 2.0, 5.0, 10.0]
print(f"\n  {'t':<8} {'d_s(t)':<12} {'Regime'}")
print(f"  {'-'*40}")
for t_val in key_t:
    idx = np.argmin(np.abs(ts - t_val))
    regime = ""
    if t_val < 0.1:
        regime = "UV (single generator)"
    elif t_val < 1:
        regime = "intermediate"
    elif t_val < 5:
        regime = "IR (group structure)"
    else:
        regime = "deep IR (trivial mode)"
    print(f"  {t_val:<8.2f} {ds_vals[idx]:<12.4f} {regime}")

# Heat kernel at t=0
Z0 = float(np.sum(np.exp(0 * eigs_flat)))
print(f"\n  Z(0) = {Z0:.0f} = |G| = phi(P4) = {PHI}")

# a_1/a_0 ratio (average Laplacian eigenvalue)
print(f"  Tr(L)/|G| = {TrL}/{PHI} = {TrL//PHI} = |S| (generator count)")
print(f"\n  The peak d_s ~ {d_peak:.2f} is between 2 and 4.")
print(f"  The solenoid is NOT a smooth manifold; d_s depends on scale.")
```
