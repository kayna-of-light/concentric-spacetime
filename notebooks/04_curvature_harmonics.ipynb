### 1. Curvature as a Function of Radius

For a circle of radius $R$, the curvature is $\kappa = 1/R$.

The **sagitta** (sag) is the maximum displacement of an arc from its chord.
For a chord of length $c$ on a circle of radius $R$:

$$s = R - \sqrt{R^2 - (c/2)^2}$$

The sagitta ratio $s/c$ measures how much the arc deviates from the chord
relative to the chord length — a direct measure of how "flat" the surface appears.

```python
# Curvature and sagitta ratio vs. radius
R = np.logspace(0, 4, 1000)  # R from 1 to 10,000
kappa = 1 / R

# Sagitta for fixed chord length c = 1
c = 1.0
sagitta = R - np.sqrt(R**2 - (c/2)**2)
sagitta_ratio = sagitta / c

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Curvature
ax1.loglog(R, kappa, color='steelblue', linewidth=2)
ax1.set_xlabel('Radius R', fontsize=12)
ax1.set_ylabel('Curvature κ = 1/R', fontsize=12)
ax1.set_title('Curvature vanishes with radius', fontsize=13)
ax1.axhline(y=0.01, color='red', linestyle='--', alpha=0.5, label='κ = 0.01')
ax1.legend()

# Sagitta ratio
ax2.loglog(R, sagitta_ratio, color='coral', linewidth=2)
ax2.set_xlabel('Radius R', fontsize=12)
ax2.set_ylabel('Sagitta ratio s/c', fontsize=12)
ax2.set_title('Arc-chord deviation (c=1) vanishes with radius', fontsize=13)
ax2.axhline(y=0.001, color='red', linestyle='--', alpha=0.5, label='s/c = 0.001')
ax2.legend()

fig.suptitle('The Linearisation of the Proprium:\nCurvature → 0 as the observer moves outward', fontsize=14, y=1.04)
plt.tight_layout()
plt.show()

# Key numerical values
print('Key values:')
for r in [1, 10, 100, 1000, 6371]:  # last = Earth radius in km
    s = r - np.sqrt(r**2 - (c/2)**2)
    print(f'  R = {r:>6}: κ = {1/r:.6f}, sagitta ratio = {s/c:.8f}')
```

### 2. Geometric Visualisation

Show how a fixed-length arc becomes indistinguishable from a straight line
as the radius of curvature increases.

```python
# Visualise arcs on circles of increasing radius
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
radii = [1, 3, 10, 50]
arc_angle = 60  # degrees

for ax, R_val in zip(axes, radii):
    # Draw the circle
    theta_full = np.linspace(0, 2*np.pi, 500)
    ax.plot(R_val * np.cos(theta_full), R_val * np.sin(theta_full), 
            color='lightgray', linewidth=0.5)
    
    # Draw the arc
    half_angle = np.radians(arc_angle / 2)
    theta_arc = np.linspace(np.pi/2 - half_angle, np.pi/2 + half_angle, 100)
    arc_x = R_val * np.cos(theta_arc)
    arc_y = R_val * np.sin(theta_arc)
    ax.plot(arc_x, arc_y, color='#E91E63', linewidth=3)
    
    # Draw the chord
    ax.plot([arc_x[0], arc_x[-1]], [arc_y[0], arc_y[-1]], 
            color='#2196F3', linewidth=2, linestyle='--')
    
    # Mark sagitta
    mid_chord_x = (arc_x[0] + arc_x[-1]) / 2
    mid_chord_y = (arc_y[0] + arc_y[-1]) / 2
    mid_arc_x = arc_x[len(arc_x)//2]
    mid_arc_y = arc_y[len(arc_y)//2]
    ax.plot([mid_chord_x, mid_arc_x], [mid_chord_y, mid_arc_y], 
            color='green', linewidth=1.5)
    
    sag = R_val - np.sqrt(R_val**2 - (R_val*np.sin(half_angle))**2)
    
    # Zoom to arc region
    margin = R_val * np.sin(half_angle) * 1.3
    ax.set_xlim(-margin, margin)
    ax.set_ylim(R_val - margin, R_val + margin*0.3)
    ax.set_aspect('equal')
    ax.set_title(f'R = {R_val}\nsag/chord = {sag/(2*R_val*np.sin(half_angle)):.4f}', fontsize=10)
    ax.set_xticks([])
    ax.set_yticks([])

fig.suptitle('Linearisation: the arc becomes the chord as R → ∞\n'
             '(red = arc, blue dashed = chord, green = sagitta)', fontsize=12, y=1.08)
plt.tight_layout()
plt.show()
```

### 3. Schwarzschild → Minkowski Transition

The Schwarzschild metric in Schwarzschild coordinates:

$$ds^2 = -\left(1 - \frac{r_s}{r}\right)c^2 dt^2 + \left(1 - \frac{r_s}{r}\right)^{-1} dr^2 + r^2 d\Omega^2$$

where $r_s = 2GM/c^2$ is the Schwarzschild radius.

As $r \to \infty$: $1 - r_s/r \to 1$, recovering Minkowski:

$$ds^2 \to -c^2 dt^2 + dr^2 + r^2 d\Omega^2$$

The curvature correction is $r_s / r$ — inversely proportional to distance.

```python
# Schwarzschild correction factor vs. r/r_s
r_over_rs = np.logspace(0.01, 4, 1000)  # from just above r_s to 10,000 r_s
correction = 1 - 1 / r_over_rs

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# g_tt component
ax1.semilogx(r_over_rs, correction, color='steelblue', linewidth=2)
ax1.axhline(y=1, color='red', linestyle='--', alpha=0.5, label='Minkowski (flat)')
ax1.set_xlabel('r / r_s', fontsize=12)
ax1.set_ylabel('$(1 - r_s/r)$', fontsize=12)
ax1.set_title('Temporal metric component\n(approaches 1 = flat)', fontsize=12)
ax1.legend()
ax1.set_ylim(-0.1, 1.1)

# Deviation from flatness
deviation = np.abs(1 - correction)
ax2.loglog(r_over_rs, deviation, color='coral', linewidth=2)
ax2.set_xlabel('r / r_s', fontsize=12)
ax2.set_ylabel('$|1 - g_{tt}/g_{tt}^{\\mathrm{flat}}|$', fontsize=12)
ax2.set_title('Deviation from Minkowski\n(curvature effect)', fontsize=12)
ax2.axhline(y=0.01, color='gray', linestyle=':', alpha=0.5, label='1% deviation')
ax2.legend()

fig.suptitle('Schwarzschild → Minkowski: Curvature vanishes with distance', fontsize=13, y=1.02)
plt.tight_layout()
plt.show()

# At what distance is the deviation negligible?
print('Deviation from flat space:')
for mult in [2, 5, 10, 100, 1000]:
    dev = 1 / mult
    print(f'  r = {mult:>5} r_s: deviation = {dev:.4f} ({dev*100:.2f}%)')
```

### 4. Spherical Harmonics: Region Count by $l$

How many distinct regions does $Y_l^m$ divide the sphere into?
This connects to the prime sequence if the region counts follow prime structure.

```python
# For zonal harmonics Y_l^0, the number of sign regions is:
# l=0: 1 region (all positive)
# l=1: 2 regions (split by equator)
# l=2: 3 regions (two cones create 3 bands)
# l=3: 4 regions (three cones create 4 bands)
# General: l+1 bands for Y_l^0

# For sectoral harmonics Y_l^l, the number of sign regions is:
# l=0: 1
# l=1: 2 (one meridional plane → 2 halves)
# l=2: 4 (two meridional planes → 4 sectors)  
# l=3: 6 (three meridional planes → 6 sectors)
# General: 2l sectors for Y_l^l (l>0)

print('Region counts for spherical harmonics:')
print('=' * 60)
print(f'{"l":>3} | {"Zonal Y_l^0":>12} | {"Sectoral Y_l^l":>14} | Thesis mapping')
print('-' * 60)

thesis_map = {
    0: 'Undivided (no prime)',
    1: 'BILATERAL → prime 2',
    2: 'VERTICAL → prime 3',
    3: 'RADIAL → prime 5 (?)',
    4: 'DEVELOPMENTAL → prime 7 (?)',
}

for l in range(5):
    zonal = l + 1
    sectoral = 2 * l if l > 0 else 1
    mapping = thesis_map.get(l, '')
    print(f'{l:>3} | {zonal:>12} | {sectoral:>14} | {mapping}')

print()
print('OBSERVATION:')
print('  Zonal regions: 1, 2, 3, 4, 5, ... = l+1')
print('  Sectoral regions: 1, 2, 4, 6, 8, ... = 2l (l>0)')
print()
print('  l=1 (2 regions) maps cleanly to prime 2 (bilateral symmetry).')
print('  l=2 (3 zonal bands) maps to prime 3 (vertical differentiation).')
print('  l=3 (4 zonal bands) does NOT map to prime 5.')
print('  The region count follows l+1, which is arithmetic, not prime.')
print()
print('  The parallel is STRUCTURAL (nesting, containment) not NUMERICAL (region count).')
print('  The thesis should not claim spherical harmonic zone counts = primes.')
```

### 5. The Cartesian Limit: Why Flatness Appears

Demonstrate: an observer on a sphere of radius $R$ who measures
distances within a patch of size $\delta \ll R$ perceives Euclidean geometry.
Deviations from flatness scale as $(\delta/R)^2$.

```python
# On a sphere of radius R, the geodesic distance between two points at 
# angular separation Δθ is d_geodesic = R * Δθ
# The chord distance is d_chord = 2R sin(Δθ/2)
# The difference: d_geodesic - d_chord ≈ R Δθ³/24 for small Δθ
# Relative error: ≈ Δθ²/24 = (d/R)²/24

delta_theta = np.logspace(-4, 0, 1000)  # angular separation
d_geodesic = delta_theta  # R=1 units
d_chord = 2 * np.sin(delta_theta / 2)
relative_error = (d_geodesic - d_chord) / d_geodesic
approx_error = delta_theta**2 / 24

fig, ax = plt.subplots(figsize=(10, 6))
ax.loglog(delta_theta, relative_error, color='steelblue', linewidth=2, label='Exact relative error')
ax.loglog(delta_theta, approx_error, color='coral', linewidth=2, linestyle='--', label=r'Approximation $\Delta\theta^2/24$')
ax.axhline(y=1e-6, color='gray', linestyle=':', alpha=0.5, label='1 ppm deviation')
ax.set_xlabel(r'Angular separation $\Delta\theta$ (radians)', fontsize=12)
ax.set_ylabel('Relative deviation from flat geometry', fontsize=12)
ax.set_title('Why the observer perceives flatness:\nCurvature deviations scale as $(\Delta\\theta)^2$', fontsize=13)
ax.legend(fontsize=11)
plt.tight_layout()
plt.show()

print('Practical flatness:')
print(f'  At Δθ = 0.01 rad (0.57°): error = {0.01**2/24:.2e}')
print(f'  At Δθ = 0.001 rad (0.057°): error = {0.001**2/24:.2e}')
print()
print('  → The observer on a large curve perceives flat Cartesian space.')
print('  → This IS the linearisation of the proprium: the centre drops')
print('     below the resolution threshold and the arc becomes a line.')
```
