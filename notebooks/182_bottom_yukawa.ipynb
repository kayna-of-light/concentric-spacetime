## 1. Algebraic Structure of 42

The ratio m_t/m_b = 42 = P₄/p₃. This is the primorial with the charge prime excised:

Every prime contributes EXCEPT p₃ = 5 (the charge prime). This means the top-bottom

```python
# Decomposition of 42
print("42 = P₄/p₃ = primorial with charge prime excised")
print(f"  P₄ = {P[4]},  p₃ = {primes[2]}")
print(f"  P₄/p₃ = {P[4]}/{primes[2]} = {P[4]//primes[2]}")
print(f"  = p₁ × p₂ × p₄ = {primes[0]} × {primes[1]} × {primes[3]} = {primes[0]*primes[1]*primes[3]}")
print()

# Role of each factor
print("Per-prime decomposition:")
print(f"  p₁ = {primes[0]}: bilateral (up/down selection)")
print(f"  p₂ = {primes[1]}: chirality (L/R structure)")
print(f"  p₄ = {primes[3]}: generation (3rd generation weight)")
print(f"  p₃ = {primes[2]}: ABSENT (charge sector cancels for isospin doublet)")
print()

# 42 as a sub-covering
# The full covering tower: S¹ ←2- S¹ ←3- S¹ ←5- S¹ ←7- S¹
# Excising p₃=5: S¹ ←2- S¹ ←3- S¹ ←7- S¹ (winding = 42)
print("Charge-neutral sub-covering: {2, 3, 7}")
print(f"  Total winding number: {primes[0]*primes[1]*primes[3]} = 42")
print(f"  Group of units: Z*₄₂")

units_42 = [k for k in range(1, 42) if np.gcd(k, 42) == 1]
phi_42 = len(units_42)
lam_42 = mlcm(mlcm(primes[0]-1, primes[1]-1), primes[3]-1)
print(f"  φ(42) = {phi_42}")
print(f"  λ(42) = lcm({primes[0]-1},{primes[1]-1},{primes[3]-1}) = {lam_42}")
print()
print(f"m_t/m_b = degree of charge-neutral sub-covering = {42}")
```

## 2. The Group Exponent Identity

The number 42 admits a second decomposition: 42 = P₃ + λ(P₄) = 30 + 12.

This is equivalent to λ(P₄) = p₁p₂(p₄ − p₃). Does this hold for other 4-prime sets?

```python

# Identity: 42 = P₃ + λ(P₄)
print(f"42 = P₃ + λ(P₄) = {P[3]} + {lam_P4} = {P[3] + lam_P4}")
print()

# Equivalent form: λ(P₄) = p₁p₂(p₄ − p₃)
lhs = lam_P4
rhs = primes[0] * primes[1] * (primes[3] - primes[2])
print(f"λ(P₄) = p₁p₂(p₄ − p₃)")
print(f"  {lhs} = {primes[0]}×{primes[1]}×({primes[3]}−{primes[2]}) = {rhs}")
print(f"  Match: {lhs == rhs}")
print()

# Proof for {2,3,5,7}:
# λ(P₄) = lcm(p₁−1, p₂−1, p₃−1, p₄−1) = lcm(1,2,4,6) = 12
# p₁p₂(p₄−p₃) = 2×3×(7−5) = 2×3×2 = 12 ✓
#
# The identity P₃ + λ(P₄) = P₄/p₃ rearranges to:
# p₁p₂p₃ + λ(P₄) = p₁p₂p₄
# → λ(P₄) = p₁p₂(p₄−p₃)  ✓

# Uniqueness test: check other 4-prime tuples
print("Uniqueness test: λ(N) vs p₁p₂(p₄−p₃)")
print(f"{'Primes':<20} {'λ(N)':<8} {'p₁p₂(p₄−p₃)':<15} {'Match':<6}")
print("-" * 50)
test_tuples = [(2,3,5,7), (2,3,5,11), (2,3,5,13), (2,3,7,11),
               (2,3,7,13), (2,5,7,11), (3,5,7,11)]
matches = []
for ps in test_tuples:
    N = ps[0] * ps[1] * ps[2] * ps[3]
    lam = 1
    for p in ps:
        lam = mlcm(lam, p - 1)
    prod = ps[0] * ps[1] * (ps[3] - ps[2])
    match = lam == prod
    mark = "✓ MATCH" if match else "✗"
    if match:
        matches.append(ps)
    print(f"  {str(ps):<18} {lam:<8} {prod:<15} {mark}")

print(f"\n→ Identity holds for {len(matches)}/{len(test_tuples)} tested tuples: {matches}")
print("  {2,3,5,7} is the first four primes; {3,5,7,11} also satisfies this form.")
print("  The identity captures consecutive-prime gap structure, not uniqueness.")
```
