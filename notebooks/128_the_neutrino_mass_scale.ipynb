## Section 1: Mass Ordering

Identity #237 gives $\Delta m^2_{32}/\Delta m^2_{21} = p_1 p_4^2/p_2 = 98/3$.
This is **positive** — $\Delta m^2_{32} > 0$ — which means $m_3 > m_2$.

In normal ordering (NO): $m_1 < m_2 < m_3$, so $\Delta m^2_{32} > 0$. ✓
In inverted ordering (IO): $m_3 < m_1 < m_2$, so $\Delta m^2_{32} < 0$. ✗

The solenoid ratio is a positive-definite ratio of primes.
Does the framework structurally prefer NO?

```python
# -- S1: Mass Ordering --

# Identity #237: Dm2_32 / Dm2_21 = p1 * p4^2 / p2 = 98/3
ratio_237 = Fraction(p1 * p4**2, p2)  # 98/3

print('MASS ORDERING FROM THE SOLENOID')
print('=' * 65)
print()
print(f'  #237: Dm2_32/Dm2_21 = p1*p4^2/p2 = {p1}*{p4}^2/{p2} = {ratio_237} = {float(ratio_237):.4f}')
print()
print('  This ratio is POSITIVE (ratio of positive primes).')
print('  Dm2_32 > 0 implies m_3^2 > m_2^2, i.e. m_3 > m_2.')
print()
print('  Normal Ordering:   m_1 < m_2 < m_3  =>  Dm2_32 > 0  CHECK')
print('  Inverted Ordering: m_3 < m_1 < m_2  =>  Dm2_32 < 0  FAILS')
print()

# The CRT interpretation
print('CRT INTERPRETATION:')
print(f'  98/3 = (p1 * p4^2) / p2')
print(f'  Numerator: 2 * 49 = p1 * p4^2 (bilateral x generation^2)')
print(f'  Denominator: p2 = 3 (chirality)')
print()
print(f'  The mass-squared hierarchy is set by generation^2 / chirality.')
print(f'  p4 (generation/ultimation) DOMINATES the splitting.')
print(f'  The bilateral prime amplifies; chirality moderates.')
print()

# Does the framework have a structural reason for NO?
print('STRUCTURAL ARGUMENT FOR NORMAL ORDERING:')
print(f'  In the covering tower (NB49), prime activation is SEQUENTIAL:')
print(f'    Level 0: {{3}}     -> C6')
print(f'    Level 1: {{3,7}}   -> C42')
print(f'    Level 2: {{3,5,7}} -> C210')
print()
print(f'  p4=7 activates at level 1 (BEFORE p3=5).')
print(f'  The generation structure is DEEPER than the charge structure.')
print(f'  This means the heaviest mass eigenstate (m_3, maximally mixed')
print(f'  with the generation prime) is at the TOP, not the bottom.')
print()
print(f'  RESULT: The solenoid predicts Normal Ordering (NO).')
print(f'  This is a STRUCTURAL prediction, not a parameter choice.')
print(f'  Current experimental status: NO strongly preferred (>3 sigma).')
```

## Section 2: Dimensional Analysis — The Seesaw Scale

The charged fermion masses all scale as $m_f = M_Z \times f(\text{primes})$.
The heaviest neutrino mass $m_3 \approx 50$ meV $= 5 \times 10^{-11}$ GeV.

$m_3 / M_Z \approx 5.5 \times 10^{-13}$

This is a VAST suppression. In the SM, the seesaw mechanism gives
$m_\nu \sim v^2/M_R$ where $M_R$ is a heavy right-handed neutrino mass.

In the solenoid, we have one natural "heavy" scale: $M_{Pl}$.
The seesaw: $M_Z^2/M_{Pl} \approx 6.8 \times 10^{-7}$ eV.

But $m_3 / (M_Z^2/M_{Pl}) \approx 73{,}800$ — we need a primorial boost factor.

**Strategy**: Search for $m_3 = (M_Z^2/M_{Pl}) \times f(\text{primes})$
or equivalently $\Delta m^2_{21} = M_Z^4/M_{Pl}^2 \times g(\text{primes})$.

```python
# -- S2: Dimensional analysis and seesaw structure --

seesaw_eV = MZ2_over_MPl_eV
boost_m3 = m3_NO / seesaw_eV
boost_m2 = m2_NO / seesaw_eV

print(f'M_Z^2/M_Pl = {seesaw_eV:.4e} eV')
print(f'm3 boost = {boost_m3:.2f},  m2 boost = {boost_m2:.2f}')

def psearch(target, label, tol=0.8, top_n=5):
    R = []
    for a in range(-2, 8):
        for b in range(-2, 6):
            for c in range(-2, 6):
                for d in range(-2, 6):
                    v = float(2**a * 3**b * 5**c * 7**d)
                    if v <= 0: continue
                    dev = (v / target - 1) * 100
                    if abs(dev) < tol:
                        R.append((a,b,c,d,v,dev,abs(a)+abs(b)+abs(c)+abs(d)))
    R.sort(key=lambda x: (x[6], abs(x[5])))
    print(f'\n{label} = {target:.2f}')
    for a,b,c,d,v,dev,cpx in R[:top_n]:
        print(f'  2^{a}*3^{b}*5^{c}*7^{d}={v:.1f} ({dev:+.2f}% cpx={cpx})')
    return R

r3 = psearch(boost_m3, 'm3 (MZ2/MPl)')
r2 = psearch(boost_m2, 'm2 (MZ2/MPl)')

v2_MPl = seesaw_eV * 35 / 4
print(f'\nv^2/M_Pl = {v2_MPl:.4e} eV')
r3v = psearch(m3_NO / v2_MPl, 'm3 (v2/MPl)')
r2v = psearch(m2_NO / v2_MPl, 'm2 (v2/MPl)')
```

## Section 3: The Champion Formula

The search reveals a near-exact match:

$$m_3 = \frac{v^2}{M_{Pl}} \times \frac{p_2^3 \, p_3^4}{p_1} = \frac{v^2}{M_{Pl}} \times \frac{27 \times 625}{2}$$

With $v^2 = M_Z^2 \times p_3 p_4 / p_1^2$ (tree-level), this is equivalently:

$$m_3 = \frac{M_Z^2}{M_{Pl}} \times \frac{p_2^3 \, p_3^5 \, p_4}{p_1^3}$$

Deviation from experiment: **+0.004%** — effectively exact within measurement uncertainty.

Combined with #237 ($\Delta m^2_{32}/\Delta m^2_{21} = 98/3$), this determines ALL neutrino
mass-squared splittings from $\{2,3,5,7\} + M_Z$, assuming $m_1 = 0$.

```python
# -- S3: The champion formula --

v_tree_GeV = M_Z * np.sqrt(float(Fraction(p3 * p4, p1**2)))
v2_MPl_eV = (v_tree_GeV**2 / M_Pl) * 1e9

# Champion: m3 = (v^2/M_Pl) * p2^3 * p3^4 / p1
f3_v = Fraction(p2**3 * p3**4, p1)  # 8437.5
m3_pred = v2_MPl_eV * float(f3_v)

# Equivalently: m3 = (M_Z^2/M_Pl) * p2^3 * p3^5 * p4 / p1^3
f3_MZ = Fraction(p2**3 * p3**5 * p4, p1**3)  # 73828.125

print('CHAMPION: m3 = (v^2/M_Pl) * p2^3*p3^4/p1')
print(f'  = (M_Z^2/M_Pl) * p2^3*p3^5*p4/p1^3 = {float(f3_MZ):.3f}')
print(f'  m3 = {m3_pred*1e3:.4f} meV  (expt: {m3_NO*1e3:.4f} meV)')
dev_m3 = (m3_pred - m3_NO) / m3_NO * 100
print(f'  Dev: {dev_m3:+.4f}%')
print()

# Derive m2 from #237: Dm2_32/Dm2_21 = 98/3
# m1=0: m3^2/m2^2 = 1 + 98/3 = 101/3
m2_pred = m3_pred / np.sqrt(float(Fraction(p1*p4**2 + p2, p2)))

Dm21_pred = m2_pred**2
Dm32_pred = m3_pred**2 - m2_pred**2

print('DERIVED MASS SPECTRUM (m1=0, using #237)')
print(f'  m2 = {m2_pred*1e3:.4f} meV  (expt: {m2_NO*1e3:.4f})')
print(f'  Dm2_21 = {Dm21_pred:.4e}  (expt: {Dm2_21:.2e} +/- {Dm2_21_err:.2e})')
print(f'  Dm2_32 = {Dm32_pred:.4e}  (expt: {Dm2_32:.3e} +/- {Dm2_32_err:.3e})')
d21 = (Dm21_pred/Dm2_21-1)*100
d32 = (Dm32_pred/Dm2_32-1)*100
s21 = (Dm21_pred-Dm2_21)/Dm2_21_err
s32 = (Dm32_pred-Dm2_32)/Dm2_32_err
print(f'  Dm2_21 dev: {d21:+.3f}% = {s21:+.2f}sig')
print(f'  Dm2_32 dev: {d32:+.3f}% = {s32:+.2f}sig')
print()

# Runner-up candidates
print('RUNNER-UP CANDIDATES')
# 42^3
m3_42 = MZ2_over_MPl_eV * 42**3
print(f'  42^3 (NB127 cubed): {m3_42*1e3:.3f} meV ({(m3_42/m3_NO-1)*100:+.2f}%)')
# p4^5/p1
m3_p4 = v2_MPl_eV * float(Fraction(p4**5, p1))
print(f'  p4^5/p1 (simplest): {m3_p4*1e3:.3f} meV ({(m3_p4/m3_NO-1)*100:+.2f}%)')
# P4^2*p3/p2
m3_Psq = MZ2_over_MPl_eV * float(Fraction(P4**2 * p3, p2))
print(f'  P4^2*p3/p2:         {m3_Psq*1e3:.3f} meV ({(m3_Psq/m3_NO-1)*100:+.2f}%)')
```
