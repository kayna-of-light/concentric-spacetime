## Part 2: Geometric Naturality — Why the Assignment is Forced

There are 4! = 24 ways to map four primes to four coordinates. We must show that only one mapping is consistent with the S² geometry and covering structure.

**Step 1**: The arena S² × R⁺ has two fundamentally different kinds of coordinates:
- **Angular** (θ, φ): live on the compact sphere S²
- **Non-angular** (r, arc): r on R⁺, arc on the covering fiber

The angular primes must be {2, 3} and the non-angular primes must be {5, 7} because:
- The smallest primes (2, 3) generate the angular S² structure (area ~ P_k²)
- The larger primes (5, 7) generate the radial shells and covering fiber

**Step 2**: Within the angular pair, φ has period 2π (full circle) and θ has range [0, π] (half-range). The bilateral cut (Z₂, p=2) naturally acts on the full circle via φ → −φ. The 3-fold stratification (celestial/spiritual/natural) naturally acts on the half-range [0, π] with three evenly spaced levels.

**Step 3**: Within the non-angular pair, r parameterizes the radial direction (R⁺) with discrete shells at P_k. The arc is the covering fiber — internal, not geometric. The middle prime p₃=5 (with φ(5)=4 charge sector) maps to the intermediate r-coordinate. The outermost prime p₄=7 (with φ(7)=6 giving maximum structure) maps to the covering fiber.

```python
# Uniqueness proof: enumerate all 24 permutations, check constraints
from itertools import permutations

coords = ['φ (full circle)', 'θ (half-range)', 'r (R⁺)', 'arc (fiber)']
angular_indices = {0, 1}  # φ, θ are angular (live on S²)
nonangl_indices = {2, 3}  # r, arc are non-angular

# Constraints:
# C1: {2,3} must map to angular coordinates (smallest primes → S²)
# C2: {5,7} must map to non-angular coordinates 
# C3: 2 → full circle (φ), not half-range (θ) [bilateral = full-circle symmetry]
# C4: 5 → r (discrete shells on R⁺), not fiber [radial structure, NB188]

n_total = 0
n_pass = 0
valid_assignments = []

for perm in permutations(range(4)):
    n_total += 1
    # perm[i] = coordinate index for prime i (p₁=2,p₂=3,p₃=5,p₄=7)
    
    # C1: primes 0,1 (=2,3) must map to angular coordinates 0,1 (=φ,θ)
    c1 = {perm[0], perm[1]} == angular_indices
    # C2: primes 2,3 (=5,7) must map to non-angular 2,3
    c2 = {perm[2], perm[3]} == nonangl_indices
    # C3: prime 0 (=2) → coordinate 0 (=φ, full circle)
    c3 = perm[0] == 0  
    # C4: prime 2 (=5) → coordinate 2 (=r)
    c4 = perm[2] == 2
    
    if c1 and c2 and c3 and c4:
        n_pass += 1
        valid_assignments.append(perm)

print("Uniqueness analysis")
print("=" * 65)
print(f"Total permutations: {n_total}")
print(f"Pass all constraints: {n_pass}")
print()
for perm in valid_assignments:
    print(f"  Valid: ", end='')
    for i, ci in enumerate(perm):
        print(f"p_{i+1}={primes[i]} → {coords[ci]}", end='  ')
    print()

# What each constraint eliminates
for label, test in [
    ("C1: {2,3} → angular", lambda p: {p[0],p[1]} == angular_indices),
    ("C1+C2: {5,7} → non-angular", lambda p: {p[0],p[1]} == angular_indices and {p[2],p[3]} == nonangl_indices),
    ("C1+C2+C3: 2 → φ (full circle)", lambda p: {p[0],p[1]} == angular_indices and {p[2],p[3]} == nonangl_indices and p[0] == 0),
    ("All: 5 → r", lambda p: {p[0],p[1]} == angular_indices and {p[2],p[3]} == nonangl_indices and p[0] == 0 and p[2] == 2),
]:
    count = sum(1 for p in permutations(range(4)) if test(p))
    print(f"  {label}: {count} survive")

print(f"\nThe assignment 2→φ, 3→θ, 5→r, 7→arc is UNIQUE (1 of 24).")
```

## Part 3: Covering Monodromy and Selection Rules on S²

On S², the p₂=3 covering creates a **branched covering** with branch points at the poles (θ = 0 and θ = π). The **monodromy** around each branch point is a Z₃ rotation of the covering sheets.

Since the monodromy acts on the azimuthal direction φ, the invariant spherical harmonics must satisfy:

$$Y_l^m(\theta, \phi + 2\pi/3) = Y_l^m(\theta, \phi)$$

This requires $e^{im \cdot 2\pi/3} = 1$, i.e., **3 | m** — the azimuthal selection rule from NB183/186.

Separately, p₁=2 provides the **bilateral symmetry** φ → −φ, which decomposes each |m| > 0 mode into cos(mφ) (even) and sin(mφ) (odd). This doesn't reduce the count but provides a parity label.

The joint action of these two primes on S² angular quantum numbers (l, m) gives:
- **m ∈ 3Z** (from p₂=3 covering monodromy)
- **m ↔ −m bilateral pairing** (from p₁=2 bilateral symmetry)
- At l = 2: only m = 0 survives → **singlet** (quarks)
- At l = 3: m ∈ {−3, 0, 3} → **triplet** (leptons: 1 + 2 modes after cos/sin split)

```python
# Selection rules from p₁=2 and p₂=3 acting on S² quantum numbers (l, m)
p1, p2 = 2, 3
l_max = 3  # double truncation from NB188

print("Joint selection rules: p₁=2 bilateral + p₂=3 covering monodromy")
print("=" * 65)
print(f"Constraint: {p2}|m (from p₂={p2} covering monodromy)")
print(f"Bilateral: m ↔ -m (from p₁={p1})")
print(f"Truncation: l ≤ {l_max}")
print()

total_modes = 0
for l in range(l_max + 1):
    m_allowed = [m for m in range(-l, l+1) if m % p2 == 0]
    n_modes = len(m_allowed)
    total_modes += n_modes
    
    # Classify
    if n_modes == 1:
        kind = "SINGLET"
    elif n_modes == 3:
        kind = "TRIPLET"
    else:
        kind = f"{n_modes}-plet"
    
    # Bilateral decomposition
    m_pos = [m for m in m_allowed if m > 0]
    m_zero = [m for m in m_allowed if m == 0]
    bilateral = f"{len(m_zero)} even(m=0)"
    if m_pos:
        bilateral += f" + {len(m_pos)} cos/sin pairs"
    
    m_str = str(m_allowed)
    print(f"  l={l}: m ∈ {m_str:>14}  →  {n_modes} mode(s)  [{kind}]  ({bilateral})")

print(f"\n  Total angular modes: {total_modes}")
print(f"  Singlet contribution (l < p₂): {sum(1 for l in range(min(p2, l_max+1)))}")
print(f"  Triplet contribution (l = p₂): {len([m for m in range(-p2, p2+1) if m % p2 == 0])}")

# The physical interpretation
print(f"\n  l=2 singlet → QUARKS (1 color-neutral state per shell)")
print(f"  l=3 triplet → LEPTONS (3 color states per shell)")
print(f"  This is the covering selection rule from NB183/186,")
print(f"  now grounded in the axis assignment: p₂=3 monodromy on φ.")
```

## Part 4: Per-Axis Metric Tensor

On S²(P_k) × R⁺ with covering fiber, the metric is diagonal in the axis coordinates:

$$ds^2 = P_k^2\,d\theta^2 + P_k^2\sin^2\theta\,d\phi^2 + dr^2 + \frac{1}{P_4}\,d(\text{arc})^2$$

Each metric component corresponds to one prime:
- **g_φφ = P_k² sin²θ** — area-weighted angular (from p₁=2)
- **g_θθ = P_k²** — pure angular (from p₂=3) 
- **g_rr = 1** — flat radial (from p₃=5)
- **g_arc = 1/P₄** — Haar fiber metric (from p₄=7, κ² = 1/P₄)

The metric range spans from g_arc = 1/P₄ (innermost, fiber) to g_angular ~ P₄² (outermost, angular), a ratio of P₄³.

```python
# Per-axis metric components
print("Per-axis metric tensor")
print("=" * 65)
print(f"{'Shell':>8} {'P_k':>6} {'g_θθ=P_k²':>12} {'g_φφ(eq)':>12} {'g_rr':>6} {'g_arc':>12} {'det(g)(eq)':>14}")
print("-" * 65)

for k in range(5):
    Pk = P[k]
    g_theta = Pk**2
    g_phi_eq = Pk**2  # at equator, sin²θ = 1
    g_r = 1
    g_arc = Fraction(1, N)
    det_g = g_theta * g_phi_eq * g_r * g_arc
    print(f"  k={k}  {Pk:>6} {g_theta:>12} {g_phi_eq:>12} {g_r:>6} {str(g_arc):>12} {str(det_g):>14}")

print(f"\nMetric scale hierarchy (at equator of outermost sphere):")
print(f"  g_angular = P₄² = {N**2}")
print(f"  g_radial  = 1")
print(f"  g_fiber   = 1/P₄ = 1/{N}")
print(f"  Range: g_angular/g_fiber = P₄³ = {N**3}")

# Volume element
print(f"\nVolume element √det(g) at equator:")
for k in range(5):
    Pk = P[k]
    sqrt_det = Fraction(Pk**2, 1) * Fraction(1, N)  # P_k⁴/P₄ under sqrt... 
    # Actually det = P_k⁴ × 1 × 1/P₄ at equator
    # √det = P_k² / √P₄
    val = Pk**2 / np.sqrt(N)
    print(f"  S²(P_{k}): √det(g) = P_{k}²/√P₄ = {Pk}²/√{N} = {val:.4f}")

print(f"\n  Volume ratio between consecutive spheres:")
for k in range(4):
    ratio = P[k+1]**2 / P[k]**2
    p_next = primes[k]
    print(f"    Vol(S²(P_{k+1}))/Vol(S²(P_{k})) = (P_{k+1}/P_{k})² = {p_next}² = {ratio:.0f}")
```

## Part 7: Factored Laplacian — Per-Axis Eigenvalue Contributions

The Laplacian on S²(P_k) decomposes into contributions from each angular axis:

$$\Delta_{S^2} = \frac{1}{P_k^2}\left[L_\theta + \frac{1}{\sin^2\theta}L_\phi\right]$$

where $L_\theta$ is the polar operator and $L_\phi = \partial^2_\phi$ is the azimuthal operator. For $Y_l^m$:
- **φ-axis** contribution: $-m^2$ (azimuthal quantum number)
- **θ-axis** contribution: $-[l(l+1) - m^2/\sin^2\theta]$ (polar, integrated)
- **r-axis** contribution: $1/P_k^2$ scaling factor (which shell)
- **arc** contribution: $\kappa^2 = 1/P_4$ damping (fiber)

The mass exponents emerge from the **innermost sphere** (P_1 = 2):
- x_lep = l(l+1)/P₁² = 3·4/4 = 3 at l=3 (NB183, #328)
- x_q = l(l+1)/P₁² × 200/189 = 2·3/4 × 200/189 = 100/63 at l=2 (NB183, #329)

```python
# Per-axis eigenvalue contributions
print("Factored Laplacian per axis")
print("=" * 65)

# For each (l, m) state allowed by covering selection:
print(f"\n  Allowed (l,m) states with 3|m and l ≤ 3:")
print(f"  {'(l,m)':<10} {'m² (φ-axis)':>14} {'l(l+1)-m² (θ)':>16} {'l(l+1)/P₁²':>14} {'Role'}")
print(f"  {'-'*60}")

P1 = P[1]  # = 2

states_by_l = {}
for l in range(l_max_val + 1):
    for m in range(-l, l+1):
        if m % p2_sel != 0:
            continue
        m2 = m**2
        ll1 = l*(l+1)
        theta_contrib = ll1 - m2  # at sin²θ = 1 (equator), this is the θ contribution
        x_val = Fraction(ll1, P1**2)
        
        if l == 2 and m == 0:
            role = "QUARK singlet"
        elif l == 3:
            role = "LEPTON triplet"
        elif l <= 1:
            role = "(sub-threshold)"
        else:
            role = ""
        
        print(f"  ({l},{m:+d})     {m2:>14} {theta_contrib:>16} {str(x_val):>14}     {role}")
        
        if l not in states_by_l:
            states_by_l[l] = []
        states_by_l[l].append((m, m2, theta_contrib))

# Mass exponents from P₁ Laplacian
print(f"\n  Mass exponents from innermost sphere (P₁ = {P1}):")
for l_check, label in [(2, "quark"), (3, "lepton")]:
    x = Fraction(l_check * (l_check + 1), P1**2)
    print(f"    x_{label} = l(l+1)/P₁² = {l_check}·{l_check+1}/{P1}² = {x} = {float(x):.6f}")

# The axis assignment determines which quantum number contributes what:
print(f"\n  Axis ↔ eigenvalue contribution:")
print(f"    φ (p₁=2): m² — azimuthal quantum number")
print(f"    θ (p₂=3): l(l+1) - m² — polar quantum number")
print(f"    r (p₃=5): 1/P_k² — shell scaling")
print(f"    arc (p₄=7): κ² = 1/P₄ — fiber damping")
print(f"  Each axis contributes independently to the total eigenvalue.")
```
